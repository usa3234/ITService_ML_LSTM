```python

from keras.models import Model
from keras.layers import Input, Dense, LSTM
import numpy as np
import pandas as pd

df = pd.read_excel('itService.xlsx')

df
```

```python
feature_df = df['MM']
label_df = df['CNT']
```

```python
xInput = Input(batch_shape=(None, 1, 1))
xLstm = LSTM(10)(xInput)
xOutput = Dense(1)(xLstm)


model = Model(xInput, xOutput)
model.compile(loss='mean_squared_error', optimizer='adam')

#print(model.summary())


model.fit(feature_df, label_df, epochs=1000, batch_size=1, verbose=1)
#model.predict(feature_df, batch_size=1)
```

```python
df_pred = pd.DataFrame({
                             'MM':[2],
                            })
model.predict(df_pred)
```
